```python
import h5py 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
folder = r"..\data\01AM"
record = "train1.hdf"
filename = os.path.join(folder, record)
with h5py.File(filename, "r") as h5f:
    data = h5f["eeg/data"][:-1]
    print(h5f.keys())
```

```python
trigger = data[:, -1]
```

```python
ttl = np.array(trigger, dtype=np.uint8)

bit = 2
photo = ((ttl>>bit) & 0b1).astype(int)

bit = 4
sound = ((ttl>>bit) & 0b1).astype(int)
```

```python
plt.plot(photo)
plt.plot(sound)
plt.xlim(230000, 250000)
```
